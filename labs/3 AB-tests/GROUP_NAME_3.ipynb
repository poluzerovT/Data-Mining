# Задание 2. Гипотезы для бинарной метрики

### Пример формулировки гипотезы:
Предположим, что каждый студент с некоторой вероятностью $p$ приходит на занятие.


Студентов можно разделить на 2 группы по признаку наличия работы. 
$X_1$ - работающие студенты
$X_2$ - не работаюище


Хотим проверить, одинаковая ли частота посещения пар среди групп. 
$H_0: {p_1 = p_2}$


Гипотеза: **Фактор "студент работает/не работает" не влияет на "посещаемость"**

---

Необходимо:
* Выдвинуть 2 гипотезы относительно влияния фактора на **бинарную** метрику
* С помощью **Биномиального теста** проверить гипотезы
* Дать текстовую интерпретацию полученных результатов
